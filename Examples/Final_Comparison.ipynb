# Final Comparison

Function initialization:

```python
from QMC_option_pricing import european_call_mc, european_call_at, european_call_ss, european_call_sobol, european_call_is

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy import stats
import time
```

## Moderate estimate

```python
S0 = 100      # spot
K = 100       # at-the-money strike
T = 1       # 1 year
r = 0.05      # 5% risk-free rate
sigma = 0.20  # 20% volatility
n = int(1e6)
exp = 100
bins = 100
```

```python
start = time.perf_counter()
prices = european_call_mc(S0, K, T, r, sigma, n, exp)
MC_time = time.perf_counter() - start
MC_var = np.var(prices, ddof=1)

start = time.perf_counter()
prices = european_call_at(S0, K, T, r, sigma, n, exp)
AT_time = time.perf_counter() - start
AT_var = np.var(prices, ddof=1)

start = time.perf_counter()
prices = european_call_ss(S0, K, T, r, sigma, n, bins, exp)
SS_time = time.perf_counter() - start
SS_var = np.var(prices, ddof=1)

start = time.perf_counter()
prices = european_call_sobol(S0, K, T, r, sigma, n, exp)
Sobol_time = time.perf_counter() - start
Sobol_var = np.var(prices, ddof=1)

MC_eff = 1/(MC_time*MC_var)
AT_eff = 1/(AT_time*AT_var)
SS_eff = 1/(SS_time*SS_var)
Sobol_eff = 1/(Sobol_time*Sobol_var)

print(f"Normalized Monte Carlo efficiency: {MC_eff/MC_eff:.0f}")
print(f"Normalized Antithetic variates efficiency: {AT_eff/MC_eff:.0f}")
print(f"Normalized Stratified sampling efficiency: {SS_eff/MC_eff:.0f}")
print(f"Normalized Sobol efficiency: {Sobol_eff/MC_eff:.0f}")
```

## Deep ITM estimate

```python
S0 = 100      # spot
K = 200       # at-the-money strike
T = 1       # 1 year
r = 0.05      # 5% risk-free rate
sigma = 0.20  # 20% volatility
n = int(1e6)
exp = 100
bins = 100
```

```python
start = time.perf_counter()
prices = european_call_mc(S0, K, T, r, sigma, n, exp)
MC_time = time.perf_counter() - start
MC_var = np.var(prices, ddof=1)

start = time.perf_counter()
prices = european_call_at(S0, K, T, r, sigma, n, exp)
AT_time = time.perf_counter() - start
AT_var = np.var(prices, ddof=1)

start = time.perf_counter()
prices = european_call_ss(S0, K, T, r, sigma, n, bins, exp)
SS_time = time.perf_counter() - start
SS_var = np.var(prices, ddof=1)

start = time.perf_counter()
prices = european_call_sobol(S0, K, T, r, sigma, n, exp)
Sobol_time = time.perf_counter() - start
Sobol_var = np.var(prices, ddof=1)

a = np.log(K/S0)/(sigma*np.sqrt(T))
start = time.perf_counter()
prices = european_call_is(S0, K, T, r, sigma, a, n, exp)
IS_time = time.perf_counter() - start
IS_var = np.var(prices, ddof=1)

MC_eff = 1/(MC_time*MC_var)
AT_eff = 1/(AT_time*AT_var)
SS_eff = 1/(SS_time*SS_var)
Sobol_eff = 1/(Sobol_time*Sobol_var)
IS_eff = 1/(IS_time*IS_var)

print(f"Normalized Monte Carlo efficiency: {MC_eff/MC_eff:.0f}")
print(f"Normalized Antithetic variates efficiency: {AT_eff/MC_eff:.0f}")
print(f"Normalized Stratified sampling efficiency: {SS_eff/MC_eff:.0f}")
print(f"Normalized Sobol efficiency: {Sobol_eff/MC_eff:.0f}")
print(f"Important sampling efficiency: {IS_eff/MC_eff:.0f}")
```
